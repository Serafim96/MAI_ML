```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


image_path_messi = '/kaggle/input/hwhwhw2/messi5.jpg'
image_messi = cv2.imread(image_path_messi)

height, width = image_messi.shape[:2]

center = (width // 2, height // 2)

angle = 180  

rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)


rotated_image = cv2.warpAffine(image_messi, rotation_matrix, (width, height))


rotated_output_path = '/kaggle/working/messi_rotated.jpg'
cv2.imwrite(rotated_output_path, rotated_image)

rotated_output_path
```

```python
image_paths = {
    "phone": "/kaggle/input/hwhwhw2/phone.png",
    "text": "/kaggle/input/hwhwhw2/text.png",
    "skull": "/kaggle/input/hwhwhw2/skull.png",
    "apple": "/kaggle/input/hwhwhw2/apple.jpg"
}

images = {name: cv2.imread(path) for name, path in image_paths.items()}

plt.figure(figsize=(10, 10))

for i, (name, image) in enumerate(images.items(), 1):
    plt.subplot(2, 2, i)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(name)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
text_image = images['text']
text_flipped = cv2.flip(text_image, 1)  

text_flipped_path = '/kaggle/working/text_flipped.png'
cv2.imwrite(text_flipped_path, text_flipped)

plt.imshow(cv2.cvtColor(text_flipped, cv2.COLOR_BGR2RGB))
plt.title('Исправленный текст')
plt.axis('off')
plt.show()

text_flipped_path
```

```python
apple_image = images['apple']
apple_flipped = cv2.flip(apple_image, 0)  

apple_flipped_path = '/kaggle/working/apple_flipped.jpg'
cv2.imwrite(apple_flipped_path, apple_flipped)


plt.imshow(cv2.cvtColor(apple_flipped, cv2.COLOR_BGR2RGB))
plt.title('Исправленное изображение с яблоком')
plt.axis('off')
plt.show()

apple_flipped_path
```

```python
image_path_phone = '/kaggle/input/hwhwhw2/phone.png'
image_phone = cv2.imread(image_path_phone)

height, width = image_phone.shape[:2]

pts1 = np.float32([[90, 250], [300, 250], [90, 330]])

pts2 = np.float32([[130, 250], [260, 250], [130, 330]])

M = cv2.getAffineTransform(pts1, pts2)

corrected_phone = cv2.warpAffine(image_phone, M, (width, height))

corrected_phone_path = '/kaggle/working/phone.png'
cv2.imwrite(corrected_phone_path, corrected_phone)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_phone, cv2.COLOR_BGR2RGB))
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(corrected_phone, cv2.COLOR_BGR2RGB))
plt.title('Скорректированное изображение')
plt.axis('off')

plt.show()

corrected_phone_path
```

```python
image_path_skull = '/kaggle/input/hwhwhw2/skull.png'
image_skull = cv2.imread(image_path_skull)

height, width = image_skull.shape[:2]

pts1 = np.float32([[119, 312], [707, 312], [174, 99], [652, 99]])

pts2 = np.float32([[119, 312], [707, 312], [119, 50], [707, 50]])

M = cv2.getPerspectiveTransform(pts1, pts2)

corrected_skull = cv2.warpPerspective(image_skull, M, (width, height))

corrected_skull_path = '/kaggle/working/skull_corrected.png'
cv2.imwrite(corrected_skull_path, corrected_skull)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_skull, cv2.COLOR_BGR2RGB))
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(corrected_skull, cv2.COLOR_BGR2RGB))
plt.title('Скорректированное изображение')
plt.axis('off')

plt.show()
```
