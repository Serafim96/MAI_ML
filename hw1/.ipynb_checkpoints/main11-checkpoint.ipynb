```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image_path = '/kaggle/input/vangog/zvezdnoe_nebo.jpeg'
image_bgr = cv2.imread(image_path)

image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)

def starry_night_to_starry_day(hsv_image):
    h, s, v = cv2.split(hsv_image)

    h = (h + 90) % 180

    v = cv2.normalize(v, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)

    s = s * 0.8
    s = np.clip(s, 0, 255).astype(hsv_image.dtype)

    adjusted_hsv = cv2.merge([h, s, v])

    return adjusted_hsv

image_day_hsv = starry_night_to_starry_day(image_hsv)

image_day_bgr = cv2.cvtColor(image_day_hsv, cv2.COLOR_HSV2BGR)

output_path = '/kaggle/working/starry_day.jpeg'
cv2.imwrite(output_path, image_day_bgr)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB))
plt.title('Original Starry Night')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(image_day_bgr, cv2.COLOR_BGR2RGB))
plt.title('Transformed Starry Day')
plt.axis('off')

plt.show()

output_path
```

```python
blur_kernel_size = (5, 5)  
image_blurred = cv2.GaussianBlur(image_bgr, blur_kernel_size, 0)

blurred_output_path = '/kaggle/working/starry_blurred.jpeg'
cv2.imwrite(blurred_output_path, image_blurred)

sharpening_kernel = np.array([[0, -1, 0],
                              [-1, 5, -1],
                              [0, -1, 0]])

image_sharpened = cv2.filter2D(image_blurred, -1, sharpening_kernel)

sharpened_output_path = '/kaggle/working/starry_sharpened.jpeg'
cv2.imwrite(sharpened_output_path, image_sharpened)

plt.figure(figsize=(15, 7))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_blurred, cv2.COLOR_BGR2RGB))
plt.title('Размытое изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(image_sharpened, cv2.COLOR_BGR2RGB))
plt.title('Увеличенное резкостью изображение')
plt.axis('off')

plt.show()

(blurred_output_path, sharpened_output_path)
```
