```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


image_path_railway = '/kaggle/input/hwhwhw3/railway.jpeg'
image_railway = cv2.imread(image_path_railway)

plt.imshow(cv2.cvtColor(image_railway, cv2.COLOR_BGR2RGB))
plt.title('Исходное изображение железной дороги')
plt.axis('on')
plt.show()
```

```python
src_points = np.float32([[200, 720], [780, 720], [120, 840], [880, 840]])
height, width = image_railway.shape[:2]

dst_points = np.float32([[130, 720], [1180, 720], [130, 1200], [1180, 1200]])

M = cv2.getPerspectiveTransform(src_points, dst_points)

corrected_railway = cv2.warpPerspective(image_railway, M, (width, height))

plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_railway, cv2.COLOR_BGR2RGB))
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(corrected_railway, cv2.COLOR_BGR2RGB))
plt.title('Преобразованное изображение')
plt.axis('off')

plt.show()
```
